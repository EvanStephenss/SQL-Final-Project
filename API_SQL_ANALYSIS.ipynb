```python
import mysql.connector
import pandas as pd

# Establishing the database connection
connection = mysql.connector.connect(
    host='isba-dev-01.crw88eg88uar.us-east-1.rds.amazonaws.com',
    user='admin',
    password='isba_4715',
    database='artworks'
)

# SQL query for individual artwork analysis
query_individual_artworks = """
SELECT
    title,
    Year,
    accessionYear,
    accessionYear - Year AS time_on_market,
    AVG(accessionYear - Year) OVER (PARTITION BY Year) AS avg_time_on_market_per_year
FROM
    roy_lichtenstein;
"""

# SQL query for yearly average time on market
query_yearly_avg_time = """
SELECT
    Year,
    AVG(accessionYear - Year) AS avg_time_on_market_per_year
FROM
    roy_lichtenstein
GROUP BY
    Year
ORDER BY
    Year;
"""

# SQL query for yearly and classification aggregates
query_yearly_classification_aggregates = """
WITH YearlyAggregates AS (
    SELECT
        Year,
        Classification,
        COUNT(*) AS NumberOfWorks,
        AVG(accessionYear - Year) AS AvgTimeOnMarket
    FROM
        roy_lichtenstein
    GROUP BY
        Year, Classification
)
SELECT
    Year,
    Classification,
    NumberOfWorks,
    AvgTimeOnMarket
FROM
    YearlyAggregates
ORDER BY
    Year, Classification;
"""

# Executing queries and fetching results
individual_artworks = pd.read_sql(query_individual_artworks, con=connection)
yearly_avg_time = pd.read_sql(query_yearly_avg_time, con=connection)
yearly_classification_aggregates = pd.read_sql(query_yearly_classification_aggregates, con=connection)

# Closing the database connection
connection.close()

# Displaying the results
print("Individual Artworks Analysis:")
print(individual_artworks)
print("\nYearly Average Time on Market:")
print(yearly_avg_time)
print("\nYearly and Classification Aggregates:")
print(yearly_classification_aggregates)
```
